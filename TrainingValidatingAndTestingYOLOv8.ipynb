## Counting the images :
Before prediction, we have a look that how many images we are going to predict our pretrained weight.

```python
import os

folder_path = "/data01/user1/FinalProject/yolov8/images/"

images = os.listdir(folder_path)

image_count = 0

for image in images:
  if image.lower().endswith(('.jpg', '.jpeg', '.gif', '.bmp')):
    image_count += 1

print("Number of images:", image_count)
```
